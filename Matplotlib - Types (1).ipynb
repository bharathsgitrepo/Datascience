```python
names = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "k", "l"]
scores = [55, 60, 65, 65, 70, 75, 80, 80, 85, 90, 90, 95, 95]

plt.bar(names, scores, color='cyan', edgecolor='black')
plt.title("Bar Chart - Student Scores")
plt.xlabel("Student")
plt.ylabel("Score")
plt.show()
```

```python
scores = [55, 60, 65, 65, 70, 75, 80, 80, 85, 90, 90, 95]
# name =["a",'b','c','d','e','f','g','h','i','j','k','k','l']
plt.hist(scores, bins=5, color='cyan',edgecolor='black')
plt.title("Histogram - Test Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Pie Chart
```

```python
activities = ['Sleep', 'Work', 'Exercise', 'Leisure']
hours = [7, 9, 2, 5]

plt.pie(hours, labels=activities)
plt.title("Pie Chart - Daily Activities") 
plt.show()
```

```python
activities = ['Sleep', 'Work', 'Exercise', 'Leisure']
hours = [8, 9, 2, 5]

plt.pie(hours, labels=activities, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart - Daily Activities")
plt.axis('equal') 
plt.show()
```

```python
# Box plot
```

```python
import matplotlib.pyplot as plt
A_scores = [70, 75, 80, 85, 90, 95, 100]
B_scores = [65, 70, 75, 80, 85, 90, 95]

data = [A_scores,B_scores]

plt.boxplot(data, labels=["Class A", "Class B"])
plt.title("Box Plot - Score Distribution")
plt.ylabel("Scores")
plt.show()
```

```python
# Stack Plot
```

```python
days = [1, 2, 3, 4, 5]
student1 = [1, 3, 4, 6, 8]
student2 = [2, 4, 5, 7, 10]

plt.stackplot(days, student1, student2, labels=["Student 1", "Student 2"], colors=['blue', 'green'])
plt.title("Area Plot - Cumulative Study Hours")
plt.xlabel("Day")
plt.ylabel("Hours")
plt.legend()
plt.show()
```

```python
# Error bar graph
```

```python
students = ["A", "B", "C", "D", "E"]
scores = [75, 85, 80, 90, 70]
errors = [2, 3, 1.5, 2.5, 3]

plt.errorbar(students, scores, yerr=errors, fmt='o', ecolor='red', capsize=5)
plt.title("Error Bar Plot - Student Scores with Uncertainty")
plt.xlabel("Student")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```
