```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

df = pd.read_csv("email_class.csv")
df['Subject'] = df['Subject'].fillna('')
df['Body'] = df['Body'].fillna('')
df['text'] = (df['Subject'] + ' ' + df['Body']).str.lower()


positive_keywords = ['we would like to offer', 'congratulations', 'selected', 'shortlisted','cleared',' make you an offer']
negative_keywords = ['not selected', 'not shortlisted', 'sorry to inform', 'unfortunately', 'rejected', 'regret', 'sorry to let', 'sorry to say','moved forward']


def assign_label(text):
    for word in negative_keywords:
        if word in text:
            return 0
    for word in positive_keywords:
        if word in text:
            return 1
    return None

df['Label'] = df['text'].apply(assign_label)
df = df.dropna(subset=['Label'])
df['Label'] = df['Label'].astype(int)
df.to_csv("email_class.csv", index=False)


X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['Label'], test_size=0.3, random_state=42
)


vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

clf = MultinomialNB()
clf.fit(X_train_tfidf, y_train)


predictions = clf.predict(X_test_tfidf)
print(classification_report(y_test, predictions))


new_email = "Interview Status Update,unfortunately your profile was not selected."
new_email_tfidf = vectorizer.transform([new_email.lower()])
result = clf.predict(new_email_tfidf)[0]
print("Prediction:", "Selected" if result == 1 else "Rejected")
```

```python
import imaplib
import email
from email.header import decode_header
import pandas as pd

# Your Gmail credentials
username = "bharathramu1310@gmail.com"
app_password = "otcu mqtn jlmd ojsv"

# Connect to Gmail IMAP server
mail = imaplib.IMAP4_SSL("imap.gmail.com")
mail.login(username, app_password)
mail.select("inbox")

# Search for all emails
status, messages = mail.search(None, 'ALL')
email_ids = messages[0].split()

emails = []

for e_id in email_ids[-20:]:  # last 20 emails
    status, msg_data = mail.fetch(e_id, "(RFC822)")
    for response_part in msg_data:
        if isinstance(response_part, tuple):
            msg = email.message_from_bytes(response_part[1])
            subject, encoding = decode_header(msg["Subject"])[0]
            if isinstance(subject, bytes):
                subject = subject.decode(encoding if encoding else "utf-8")
            # Extract email body (plain text)
            body = ""
            if msg.is_multipart():
                for part in msg.walk():
                    ctype = part.get_content_type()
                    cdispo = str(part.get("Content-Disposition"))
                    if ctype == "text/plain" and "attachment" not in cdispo:
                        try:
                            body = part.get_payload(decode=True).decode()
                        except:
                            pass
                        break
            else:
                body = msg.get_payload(decode=True).decode()

            emails.append({"Subject": subject, "Body": body})

mail.logout()

# Create DataFrame and save CSV
df = pd.DataFrame(emails)
print(df.head())
df.to_csv("imap_fetched_emails.csv", index=False)
```
