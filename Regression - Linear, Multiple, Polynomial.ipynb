```python

experience = [1, 2, 3, 4, 5]
salary = [30000, 35000, 40000, 45000, 50000]

# Step 2: Calculate means
n = len(experience)
x1 = sum(experience) / n
y1 = sum(salary) / n

# y =mx+c
# m =sum((x-x1)(y-y1))/sum((x-x1)**2)

a = 0
for i in range(n):
    a+= (experience[i] - x1) * (salary[i] - y1)

    
b =0
for i in range(n):
    b+= (experience[i] - x1)**2


m = a/b

c = y1-m* x1 

print(f"The linear regression equation is: y = {m:.2f}x + {c:.2f}")


x_input = float(input("Enter years of experience: "))

predicted_salary = m * x_input + c
print(f"Predicted Salary for {x_input} years of experience is: {predicted_salary:.2f}")
```

```python
x = [1, 2, 3, 4, 5]
y = [30000, 35000, 40000, 45000, 50000]

m = 0
c = 0
n = len(x)
learning_rate = 0.01
epochs = 1000


for _ in range(epochs):
    y_pred = [m * x[i] + c for i in range(n)]
    dm = (-2/n) * sum(x[i] * (y[i] - y_pred[i]) for i in range(n))
    dc = (-2/n) * sum(y[i] - y_pred[i] for i in range(n))
    m -= learning_rate * dm
    c -= learning_rate * dc
print(dm)
print(dc)
# print(m)
# print(c)
print(f"Final model: y = {m:.2f}x + {c:.2f}")

x_input = float(input("Enter years of experience: "))
predicted_salary = m * x_input + c
print(f"Predicted Salary for {x_input} years of experience: {predicted_salary:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures


x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25])  

poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
model = LinearRegression()
model.fit(x_poly, y)
x_range = np.linspace(1, 5, 100).reshape(-1, 1)
x_range_poly = poly.transform(x_range)
y_pred = model.predict(x_range_poly)

plt.scatter(x, y, color='blue', label='Original Data')
plt.plot(x_range, y_pred, color='red', label='Polynomial Fit (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```
